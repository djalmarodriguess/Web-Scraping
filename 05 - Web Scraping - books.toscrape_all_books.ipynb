# Como fazer WEBSCRAPING utilizando Beautiful Soup

# Importar Bibliotecas

```python
# Importar Bibliotecas
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Coleta dos HTMLS

```python
# Entrar com a url e headers

headers = {'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36'}

html_list = []
for i in range(1,51):
    print('Page {}'.format(i))
        
    url = 'https://books.toscrape.com/catalogue/page-{}.html'.format(i) # Link da página web a ser analisada
    
    print(url)
    
    # Fazer request
    response = requests.get(url = url, headers = headers)
    
    if response.status_code == 200:
        
        print('Solicitação bem-sucedida')
        html_page = response.text

        html_list.append(html_page)
        
    else:
        print('Solicitação mal-sucedida')

```

# Encontrar os Títulos

```python
titleList = []
for i in html_list:    
    
    soup = BeautifulSoup(i,'html.parser')

    titleSoup = soup.find_all('a')

    # Lista com os títulos
    # Valores da localização vão de 2 em 2 a partir do 54º posição da tag </a>
    title = [j.get('title') for i, j in enumerate(titleSoup) if i >=54 and i %2 == 0 and j.get('title') is not None]

    # Mostrar Título
    titleList.extend(title)
    
titleList
```


# Encontrar Preço

```python
priceList = []

for i in html_list:
    # Criar variavel do preço
    priceSoup = soup.find_all('p', class_="price_color")

    # Buscar todos os preços
    price = [float(i.string.replace('Â£', '')) for i in priceSoup] # Colocar formato Float e substituir caracter indevido
    
    priceList.extend(price)

# Mostrar preço
priceList
```

# Encontrar Avaliações dos Livros

```python
starratingList = []
for i in html_list:
    
    # Chamar todas as avaliações
    starSoup = soup.find_all('p', class_='star-rating')

    # Criar variavel
    starrating = [rating.get('class')[1] for rating in starSoup]
    
    starratingList.extend(starrating)

starratingList
```

# Fazer DataFrame

```python
# O tamanho das variavel devem ser sempre o mesmo
print(len(titleList))
print(len(priceList))
print(len(starratingList))
```

```python
# Dicionario com as listas
data ={ 'title' : titleList,
        'price' : priceList ,
        'star_rating': starratingList
      }


# DataFrame
df = pd.DataFrame(data)
df
```

```python
# Criar coluna para classificação numerica das avaliações
df['star_rating_numeric'] = df['star_rating'].map({'One': 1, 'Two': 2, 'Three': 3, 'Four': 4, 'Five': 5})
df.head()
```

# Plots

```python
# Criar figure e axes
fig,ax = plt.subplots(1,2,figsize=(10,5))

# Plotar avaliações de cada Livro
ax[0].boxplot(df.price)
ax[0].set_title('Distribuição preço livro'  )

# Plotar distribuição das avaliações
ax[1].hist(df['star_rating_numeric'], color = 'skyblue')
ax[1].set_title('Distribuição das avaliações')

# Mostrar subplots
plt.show()
```
