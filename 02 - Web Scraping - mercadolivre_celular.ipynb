# Importar Bibliotecas


```python
# Importar Bibliotecas
import requests
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt
```

# Requisição na url

```python
# Entrar com a url e headers
url = 'https://www.mercadolivre.com.br/ofertas?container_id=MLB779362-4&category=MLB1055&promotion_type=deal_of_the_day#deal_print_id=bbfbabe0-b48f-11ee-a1d6-818b956a69b2&c_id=carouseldynamic-normal&c_element_order=undefined&c_campaign=VER-MAIS&c_uid=bbfbabe0-b48f-11ee-a1d6-818b956a69b2' # Link da página web a ser analisada
headers = {'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36'}

# Fazer request
page = requests.get(url, headers)
page
```

```python
soup = BeautifulSoup(page.text, 'html.parser')
soup
```

# Extrair nome do produto

```python
nomeSoup = soup.find_all('p', class_="promotion-item__title")

nomeList = [j.text for i, j in enumerate(nomeSoup)]
nomeList
```

# Extrair preço produto

```python
soup.find_all('span')
```

```python
soup.find_all('span', class_='andes-money-amount__fraction')
```

```python
soup.find_all('span', class_='andes-money-amount__fraction')[0]
```

```python
precoSoup = soup.find_all('span', class_='andes-money-amount__fraction')
precoList = [j.text.replace('.', ',') for i, j in enumerate(precoSoup) if i >= 0 and i % 2 == 0]
precoList
```

# Extrair desconto descrito no site

```python
soup.find_all('span', class_="promotion-item__discount-text")
```

```python
decontoSoup = soup.find_all('span', class_="promotion-item__discount-text")

descontoList = [int(j.text.split()[0].replace('%', '')) if j.text.strip() else None for j in decontoSoup]
descontoList
```

# Extrair preço sem desconto descrito no site

```python
soup.find_all('span', class_="andes-money-amount__fraction")
```

```python
soup.find_all('span', class_="andes-money-amount__fraction")[1].string
```

```python
valorrealSoup = soup.find_all('span', class_="andes-money-amount__fraction")

valorrealList = [j.text.replace('.', ',') for i, j in enumerate(valorrealSoup) if i >= 1 and i % 2 != 0]
valorrealList
```

# DataFrame

```python
print(len(nomeList))
print(len(valorrealList))
print(len(descontoList))
print(len(precoList))
```

```python
data = { 'produto':nomeList,
         'valor':valorrealList,
         'desconto_no_site':descontoList,
         'preco_final':precoList
      }

df = pd.DataFrame(data)
df
```
